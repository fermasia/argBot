```python
import pandas as pd
import numpy as np
```

```python
# Cargar tabla de tweets de usuarios con Hashtags Macristas
tweets_k = pd.read_pickle('../datasets/TweetsDataset_K.pkl')
tweets_k['grupo'] = 'K'
tweets_m = pd.read_pickle('../datasets/TweetsDataset_M.pkl')
tweets_m['grupo'] = 'M'
#tweets = pd.concat([tweets_k,tweets_m])
```

```python
# Quedarnos solo con el texto de cada tweet y transformarlo en lista
textos_k = tweets_k.text.tolist()
textos_m = tweets_m.text.tolist()
```

```python
#STOPWORDS
import nltk
from nltk.corpus import stopwords
stopwords.words('spanish')
#from stop_words import get_stop_words
stopwords_extra = ['bit','bitly', 'www','twitter','com','https','status','xa0','pic','xa0pic','Hoy','Ahora','si',"xA0'","año","día","así"]
stopwords = stopwords.words('spanish') + stopwords_extra
```

```python
# Remove Stopwords
textos_k = [word for word in textos_k if word not in stopwords]
textos_m = [word for word in textos_m if word not in stopwords]
```

```python
#Wordcloud M
from wordcloud import WordCloud
from stop_words import get_stop_words
import matplotlib.pyplot as plt
text = textos_m
wordcloud = WordCloud(
   width = 1500,
   height = 2000,
   background_color = 'white',
   stopwords = stopwords).generate(str(text))
fig = plt.figure(
   figsize = (30, 20),
   facecolor = 'w',
   edgecolor = 'w')
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```
