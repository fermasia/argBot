```python
import pandas as pd
import numpy as np
from IPython.display import HTML, display
```

```python
users_ss1 = pd.read_csv('trainDatasets/users_ss1.csv')
users_ss2 = pd.read_csv('trainDatasets/users_ss2.csv')
users_ss3 = pd.read_csv('trainDatasets/users_ss3.csv')
users_humans = pd.read_csv('trainDatasets/users_gen.csv')
users_humans.head(2)
```

```python
users_ss1.loc[:,'bot'] = True
users_ss2.loc[:,'bot'] = True
users_ss3.loc[:,'bot'] = True
users_humans.loc[:,'bot'] = False

users_ss1.loc[:,'user_type'] = 'Spambot_1'
users_ss2.loc[:,'user_type'] = 'Spambot_2'
users_ss3.loc[:,'user_type'] = 'Spambot_3'
users_humans.loc[:,'user_type'] = 'Human'
```

```python
users = pd.concat([users_ss1,users_ss2,users_ss3,users_humans],ignore_index=True)
users.head()
```

```python
[x for x in zip(users.columns, users.iloc[0,:].values)]
```

```python
users = users[[
    'id', #...id
    'screen_name', #@user
    'name', #complete name
    'description',
    'verified',
    'crawled_at',
    'location',
    'url',
    'created_at',
    'statuses_count',
    'listed_count',
    'favourites_count',
    'followers_count',
    'friends_count',
    'bot',
    'user_type'
]]

users.columns = [
    'id',
    'screen_name',
    'complete_name',
    'description',
    'verified',
    'crawled_at',
    'location',
    'url',
    'created_at',
    'tweet_count',
    'listed_count',
    'favs_count',
    'followers_count',
    'following_count',
    'bot',
    'user_type'
]
users.index = users.id.values
```
